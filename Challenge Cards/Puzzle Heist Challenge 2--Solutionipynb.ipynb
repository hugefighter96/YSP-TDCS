## Puzzle Heist: Challenge 2

### Difficulty Level: Hard

### Instructions
Doing this challenge **requires** you to have done the Puzzle Heist: Challenge 1. If you have completed that, then write a python program for the entire puzzle that outputs the final 4-digit number.


```python
word = "mAkErSpAcE"
letter7 = word[6]
letter = letter7.lower()
vowels = "aeiou"
unique_vowels = ""
for ch in word:
    c = ch.lower()
    if c in vowels and c not in unique_vowels:
        unique_vowels = unique_vowels + c
vowel_count = 0
for _ in unique_vowels:
    vowel_count += 1
shifted_ascii = ord(letter) + vowel_count
if shifted_ascii > ord('z'):
    shifted_ascii = ord('a') + (shifted_ascii - ord('z') - 1)
shifted_char = chr(shifted_ascii)
unique_consonants = ""
for ch in word:
    c = ch.lower()
    if c.isalpha() and c not in vowels and c not in unique_consonants:
        unique_consonants = unique_consonants + c
consonant_count = 0
for _ in unique_consonants:
    consonant_count += 1
total = ord(shifted_char) + consonant_count
while total >= 10:
    sum_digits = 0
    for ch in str(total):
        sum_digits += int(ch)
    total = sum_digits
digit1 = total

sentence = "Cyber security is a shared responsibility, not just an IT concern."
sentence = sentence.replace(",", "").replace(".", "")
words = sentence.split()
m = 0
for word in words:
    if len(word) < 5:
        m = m + 1
n_words = 0
for word in words:
    if len(word) > 5:
        n_words += 1
n = n_words + 1
nth = words[n - 1]
rev_nth = ""
for i in range(len(nth)-1, -1, -1):
    rev_nth += nth[i]
mth = words[m - 1]
no_vowels = ""
for ch in mth:
    if ch.lower() not in vowels:
        no_vowels += ch
new_word = rev_nth + no_vowels
count_letters = 0
for ch in new_word:
    if ch.isalpha():
        count_letters += 1
while count_letters >= 10:
    sum_digits = 0
    for ch in str(count_letters):
        sum_digits += int(ch)
    count_letters = sum_digits
digit2 = count_letters

grid = [
    ['T', 'R', 'A', 'C', 'E'],
    ['H', 'A', 'C', 'K', 'S'],
    ['O', 'N', 'L', 'I', 'N'],
    ['E', 'D', 'A', 'T', 'A'],
    ['P', 'A', 'S', 'S', 'W'],
    ['O', 'R', 'D', 'S', '!']
]
coords = [(0,0), (1,2), (2,3), (3,1), (4,4), (5,0), (2,0), (3,3), (5,2)]
letters = ""
for pair in coords:
    r = pair[0]
    c = pair[1]
    letters += grid[r][c]
rev = ""
for i in range(len(letters)-1, -1, -1):
    rev += letters[i]
unique_cons = ""
for ch in rev:
    c = ch.lower()
    if c not in vowels and c.isalpha() and c not in unique_cons:
        unique_cons += c
count = 0
for _ in unique_cons:
    count += 1
while count >= 10:
    sum_digits = 0
    for ch in str(count):
        sum_digits += int(ch)
    count = sum_digits
digit3 = count

letters = ['B', 'E', 'H', 'K', 'N']
positions = []
for i in range(5):
    p = ord(letters[i]) - ord('A') + 1
    positions.append(p)
diff = positions[1] - positions[0]
next_pos = positions[4] + diff
x = next_pos
y = 5
z = y % x
digit4 = z

code = str(digit1) + str(digit2) + str(digit3) + str(digit4)
print("Final Heist Code:", code)
```
